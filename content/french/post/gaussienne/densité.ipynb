# Les math√©matiques symboliques avec Python

## Fonction de densit√© d'une gaussienne


```python
from sympy import symbols, pi, exp, sqrt, plot
x, mu, sigma = symbols("x, mu, sigma")
```

```python
density = (1/(sigma*sqrt(2*pi))) * (exp(-1/2* ((x-mu)/sigma)**2))
```

## Repr√©sentation graphique d'une gaussienne centr√©e r√©duite

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("ggplot")
plt.rcParams["figure.figsize"] = (8, 2)
plt.rcParams["figure.dpi"] = 300
plot(density.subs({mu : 0, sigma : 1}))
```

## Repr√©sentation de sa d√©riv√©e aux points {mu : 0, sigma :1)

```python
plot(density.diff(x).subs({mu : 0, sigma : 1}))
```
